## Linear Regression

```python
n = 30
X = np.linspace(0,1,n)
y = X + np.random.normal(0,0.2,n)
```

```python
model = sklearn.linear_model.LinearRegression().fit(X.reshape(-1,1),y)
```

```python
X_pred = np.linspace(np.min(X),np.max(X),10*n)
y_pred = model.predict(X_pred.reshape(-1,1))
```

```python
plt.scatter(X,y)
plt.plot(X_pred,y_pred,c='r')
plt.axis('off')
plt.savefig("static/linear_regression.png",bbox_inches='tight')
```

## Linear Discriminator

```python
n = 30
X1, y1 = np.random.multivariate_normal([1,1],[[0.5,0],[0,0.5]],int(n/2)).T
X2, y2 = np.random.multivariate_normal([-1,-1],[[0.2,0],[0,0.2]],int(n/2)).T
```

```python
X = np.concatenate((X1,X2))
y = np.concatenate((y1,y2))
labels = np.concatenate((np.ones(int(n/2)), np.zeros(int(n/2))))
```

```python
model = sklearn.linear_model.LogisticRegression().fit(np.stack((X,y)).T, labels)
```

```python
X_pred = np.linspace(np.min(X),np.max(X),10*n)
y_pred = (0.5 - model.intercept_ - model.coef_[0,0]*X_pred)/model.coef_[0,1]
```

```python
plt.scatter(X1,y1,c='b',marker='_')
plt.scatter(X2,y2,c='r',marker='+')
plt.plot(X_pred,y_pred,c='k')
plt.axis('off')
plt.savefig("static/logistic_regression.png",bbox_inches='tight')
```

## Clustering

```python
n = 30
X1, y1 = np.random.multivariate_normal([1,1],[[0.05,0],[0,0.05]],int(n/3)).T
X2, y2 = np.random.multivariate_normal([0,1- np.sin(15*np.pi/180) * 2/3**0.5],[[0.05,0],[0,0.05]],int(n/3)).T
X3, y3 = np.random.multivariate_normal([1- np.sin(15*np.pi/180) * 2/3**0.5,0],[[0.05,0],[0,0.05]],int(n/3)).T
```

```python
X = np.concatenate((X1,X2,X3))
y = np.concatenate((y1,y2,y3))
```

```python
model = sklearn.cluster.KMeans(3).fit(np.stack((X,y)).T)
```

```python
vor = scipy.spatial.Voronoi(model.cluster_centers_)
```

```python
fig,ax = plt.subplots(1)
ax.scatter(X1,y1,c='r')
ax.scatter(X2,y2,c='r')
ax.scatter(X3,y3,c='r')
xlim = ax.xaxis.get_data_interval()
mxlim = np.mean(xlim)
xlim = (xlim - mxlim)*1.1 + mxlim
ylim = ax.yaxis.get_data_interval()
mylim = np.mean(ylim)
ylim = (ylim - mylim)*1.1 + mylim
scipy.spatial.voronoi_plot_2d(vor,ax,show_points=False,show_vertices=False)
ax.set_xlim(*xlim)
ax.set_ylim(*ylim)
ax.axis('off')
fig.savefig('static/kmeans_clustering.png',bbox_inches='tight')
```

## LOESS model parameters edge detector

```python
import local_models.local_models
import scipy.stats
```

```python
X = np.arange(n)
y = scipy.stats.logistic(n/2,n/20).cdf(X) + np.random.normal(0,0.05,n)
```

```python
plt.scatter(X,y)
```

```python
models = local_models.local_models.LocalModels(sklearn.linear_model.LinearRegression())
```

```python
models.fit(X.reshape(-1,1),y)
```

```python
kernel = local_models.local_models.TriCubeKernel(bandwidth=n/5)
X_pred = np.linspace(np.min(X), np.max(X), 5*n)
y_pred = models.predict(X_pred.reshape(-1,1), weighted=True, kernel=kernel, r=kernel.support_radius())
```

```python
y_params = models.transform(X_pred.reshape(-1,1), weighted=True, kernel=kernel, r=kernel.support_radius())
```

```python
dat = plt.scatter(X, y, c='b')
loess = plt.plot(X_pred,y_pred,c='r')
loess_slope = plt.plot(X_pred,y_params[:,0]/np.max(y_params[:,0]),c='k')
plt.legend([dat, loess[0], loess_slope[0]], ["data", "LOESS", "LOESS slope parameter"])
plt.axis("off")
plt.savefig("static/logistic_data_loess_and_params.png",bbox_inches='tight')
```

```python
y_as_image = cv2.normalize(
    np.tile(y,(n,1)),
    None,0,255,cv2.NORM_MINMAX)
loess_as_image = cv2.normalize(
    np.tile(y_pred,(y_pred.shape[0],1)),
    None,0,255,cv2.NORM_MINMAX)
loess_params_as_image = cv2.normalize(
    np.tile(y_params[:,0], (y_params.shape[0],1)),
    None,0,255,cv2.NORM_MINMAX)
```

```python
plt.imshow(y_as_image,cmap=cm.gray)
plt.axis("off")
cv2.imwrite("static/logistic_data_as_image.png", 
    cv2.resize(y_as_image, (500,500), interpolation=cv2.INTER_NEAREST)
    )
```

```python
plt.imshow(loess_as_image,cmap=cm.gray)
plt.axis("off")
cv2.imwrite("static/logistic_data_loess_as_image.png", cv2.resize(loess_as_image,(500,500),interpolation=cv2.INTER_NEAREST))
```

```python
plt.imshow(loess_params_as_image,cmap=cm.gray)
plt.axis("off")
cv2.imwrite("static/logistic_data_loess_params_as_image.png", cv2.resize(loess_params_as_image,(500,500),interpolation=cv2.INTER_NEAREST))
```
