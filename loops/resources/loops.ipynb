# For loops

### Recipe:

```python
for each_item in some_iterable:
    do_stuff(each_item)
```

````python
for i in range(2,19,3):
    print(i)
````

### Accumulator:

````python
total = 0
for i in range(2,19,3):
    total += i
print(total)
````

````python
total = 0
for each_item in [5,9,8]:
    total += each_item
print(total/3)
````

````python
my_set = {"pickles",1,2,3,5,8}
for each_item in my_set:
    print(each_item)
````

````python
my_name = "scott brown"
for each_char in my_name:
    print(each_char)
````

# While loops

### Recipe:

```python
while boolean_condition:
    do_something()
```

````python
i = 0
while i < len(my_name):
    print(my_name[i])
    i += 1
````

### Looping until condition vs looping over everything

````python
while True:
    print("hello world")
````

````python
while False:
    print("goodbye, cruel world!")
````

````python
# Save your work at this point for this to work
this_file = open("loops.ipynb")
i = 0
a_line_of_this_file = this_file.readline()
while "hello" not in a_line_of_this_file:
    i += 1
    a_line_of_this_file = this_file.readline()
print(i, a_line_of_this_file)
````

````python
this_file = open("loops.ipynb")
j = 0
for line in this_file:
    if "hello" in line:
        break # don't do this unless really necessary
    j += 1
print(j, line)
````

# Nested Loops

### Recipe:

```python
for each_item1 in any_iterable1:
    for each_item2 in any_iterable2:
        do_something(each_item1, each_item2)
```

````python
chess_columns = 'abcdefgh'
````

````python
for each_row in range(1,9):
    for each_column in chess_columns:
        print(each_column + str(each_row))
````
